```python
sigmaobs_performance = df.groupby(['sigmaobs']).mean().drop(columns=['RSEED'])
```

```python
sigmaobs_performance.index
```

```python
plt.plot(sigmaobs_performance.index,sigmaobs_performance.loc[:,'acc'])
```

```python
sigmaobs_variance = df.drop(columns=['RSEED']).groupby(['sigmaobs']).var()
```

```python
plt.errorbar(sigmaobs_performance.index,sigmaobs_performance.loc[:,'mae'],yerr=sigmaobs_variance.loc[:,'mae'])
```

```python
dfDA = pd.read_csv('DA_temp.csv')
```

```python
dfDA
```

```python
dfDA.drop(columns=['data_dir','model_dir','RSEED','sigmaobs','nobs','deltaobs','sigmainit','nens'],inplace=True)
```

```python
dfDA2 = dfDA.groupby(by=['ferr','infl_nlin']).mean()
```

```python
dfDA2
```

```python
ferr_vals = dfDA2.index.levels[0]
```

```python
fr1 = dfDA2.xs(0.1,level='ferr',axis=0)
```

```python
fr1
```

```python
plt.plot(fr1)
```

```python
k = 1
plt.figure(figsize=(20,20))
for i in ferr_vals:
    arr = dfDA2.xs(i,level='ferr',axis=0)
    plt.subplot(7,3,k)
    k+=1
    plt.plot(arr)
    plt.title('ferr={:f}'.format(i))
```
