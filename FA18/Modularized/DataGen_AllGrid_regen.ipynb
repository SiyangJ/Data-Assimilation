```python
import os
import sys
import numpy as np
import pandas as pd
```

```python
#data_dir = '/Volumes/BOY/DAmissingdata/'
data_dir = '/Volumes/BOY/Data-Assimilation/'
new_data_dir = '/Volumes/BOY/DANewData/'
```

```python
IndexList = [[fc for fc in np.linspace(-10,100,12)],
             [am for am in np.linspace(0.2,0.8,31)],
             [E for E in np.linspace(-200,0,41)],
             ['State']+[sigma for sigma in np.linspace(0,5,6)]]

IndexNames = ['fc','am','E','sigma']
```

```python
DataIndex = pd.MultiIndex.from_product(IndexList)
```

```python
DataIndex.rename(IndexNames,inplace=True)
```

```python
DF = pd.DataFrame(columns=DataIndex)
```

```python
for fc in np.linspace(-10,100,12):
    config.ResetValue('TrueModel','Fc',str(fc))
    config.ResetValue('RunModel','Fc',str(fc))
    for am in np.linspace(0.2,0.8,31):
        for E in np.linspace(-200,0,41):
            xinit_string = '{:.5f},{:.7f}'.format(E,am)
            data_save_dir = data_dir+'S_fc={:d}_am0={:4.2f}_E0={:d}.npz'.format(int(fc),am,int(E))
            _d = np.load(data_save_dir)
            DF[fc][am][E]['State'] = [_d['truestate'].astype(np.half)]
            _d.close()
            for sigma in np.linspace(0,5,6):
                obs_save_dir = data_dir+'O_fc={:d}_am0={:4.2f}_E0={:d}_sigma={:d}.npz'.format(int(fc),am,int(E),int(sigma))
                _s = np.load(obs_save_dir)
                DF[fc][am][E][sigma] = [_s['output_observation'].astype(np.half)]
                _s.close()
        print('Finish ',fc,xinit_string)
        
```
